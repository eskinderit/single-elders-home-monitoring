```python
def sq_sum(x):
  return sum((y ** 2 for y in x))

def sq_dist(x, y):
  return sq_sum((z[0] - z[1] for z in zip(x, y)))
```

```python
input_file = sc.textFile('spline_boundary_enum_inputs.csv')
```

```python
type(input_file)
```

Collect shows the content of an RDD. Note that an RDD is a sequence.

```python
input_file.collect()
```

We separate each string in the RDD into the components of a triple, converting to numbers

```python
patterns = input_file.map(lambda line: [int(line.split(',')[0])] + [float(x) for x in line.split(',')[1:]])
print(type(patterns))
patterns.collect()
```

```python
distances = patterns.map(lambda pattern: [pattern[0], sq_dist(pattern[1:], (0,0))])
print(type(distances))
distances.collect()
```

Compute the closest object by reducing through the associative and commutative min function.  Specify the order in the key argument as the second element of each item

```python
nearest = distances.reduce(lambda a, b: min(a, b, key=lambda k: k[-1]))
print(type(nearest))
print(nearest)
```
