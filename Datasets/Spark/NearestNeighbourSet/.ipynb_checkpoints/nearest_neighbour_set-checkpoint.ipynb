```python
import csv
```

```python
def sq_sum(x):
  return sum((y ** 2 for y in x))

def sq_dist(x, y):
  return sq_sum((z[0] - z[1] for z in zip(x, y)))
```

```python
query_file = "query_points.csv"  # name of query file
#query_file = str(sys.argv[1])                  # for batch mode
with open(query_file, newline='') as csvfile:
  reader = csv.reader(csvfile, delimiter=',')
  queries = []
  q_id = 0
  for q in reader:                              # fill list with points in file
    queries.append([q_id] + [float(x) for x in q[1:]])
    q_id += 1
```

```python
input_file = sc.textFile('spline_boundary_enum_inputs.csv')
```

We separate each string in the RDD into the components of a triple, converting to numbers

```python
patterns = input_file.map(lambda line: [int(line.split(',')[0])] + [float(x) for x in line.split(',')[1:]])
```

Generate distance values for each pattern and query point, placing the query id in first position as the key for a subsequent reduction.  The value is the pair pattern id, distance.  Note that the query point distances to a pattern are created as a list of lists by the lambda function, and would thus be collected as a list of lists of lists in the output RDD.  To eliminate one level, use flatMap.

```python
distances_queries = patterns.flatMap(lambda pattern: [[q[0], [pattern[0], sq_dist(pattern[1:], q[1:])]] for q in queries])
print(type(distances_queries))
distances_queries.collect()
```

Compute the closest object for each query id by reducing through the associative and commutative min function over distances.  Specify the order in the key argument as the second element of each item.  Note that the argument function in reduceByKey expects to be applied to values

```python
nearest = distances_queries.reduceByKey(lambda a, b: min(a, b, key=lambda k: k[-1]))
print(type(nearest))
print(nearest)
nearest.collect()
```
